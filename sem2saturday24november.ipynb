```python
#2.	Загрузив текст из этого файла, сделать полный грамматический разбор и записать результаты в файл json.
import json

with open('Neznakomka.txt', 'r', encoding='cp1251') as f:
            poem = f.read()
            poema = m.analyze(poem)

with open('poema.json', 'w', encoding='cp1251') as f:
    json.dump(poema, f, ensure_ascii=False)
    
#проверяем грамматический разбор    
with open('poema.json', 'r', encoding='cp1251') as f:
    outcome = f.read()
    print(outcome)
       
    
    
```

```python
#3.	Написать регулярное выражение, которое будет извлекать из тега только часть речи.
words = {}
for word in poema:
    try:
        form = word['text']
        gr = word['analysis'][0]['gr']
        words[form] = gr
    except (KeyError, IndexError) as e :
        pass
    
print(words)
```

```python
import re
pos = re.compile('[A-Z]+')
words_pos = {}
for k, v in words.items():
    words_pos[k] = pos.match(v).group(0)
print(words_pos)
```
