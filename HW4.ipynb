2.	Очистите текст от пунктуации, приведите его к нижнему регистру, лемматизируйте и запишите в новый файл. Проверьте, выполняется ли здесь закон Ципфа. Для этого постройте линейную диаграмму, где по оси Х будет ранг (порядковый номер) слова в отсортированном частотном словаре, а по оси Y - относительная частота (не ipm, а просто абсолютная частота, разделенная на количество слов в тексте).

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import WordNetLemmatizer
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
from pymystem3 import Mystem
from collections import Counter
```

```python
with open('Harry_P.txt', 'r', encoding='utf-8') as f:
    novel = f.read()
    
# очищаем текст от пунктуации

no_punkt = [w.strip("!\"#$%&'()*+,./:;<=>?@[\]^_`{|}~„“«»†*—/\-—") for w in novel.split()]

# очищаем от пустых элементов

no_punkt = [w for w in no_punkt if len(w) != 0]
   
# приводим к нижнему регистру

no_punkt = ' '.join(no_punkt).lower()
```

```python
no_punkt[:200]
```

```python
# лемматизируем и записываем в новый файл

from nltk import WordNetLemmatizer

wnl = WordNetLemmatizer()
harry_lemmatized = ' '.join((wnl.lemmatize(w) for w in no_punkt.split()))

with open('harry_lemmatized.txt', 'w', encoding='utf-8') as f:
    f.write(harry_lemmatized)
```

Проверьте, выполняется ли здесь закон Ципфа. Для этого постройте линейную диаграмму, где по оси Х будет ранг (порядковый номер) слова в отсортированном частотном словаре, а по оси Y - относительная частота (не ipm, а просто абсолютная частота, разделенная на количество слов в тексте).

```python
# cчитаем абсолютную частоту всех лемм
harry_freq = {}
counts = Counter(harry_lemmatized.split())

for k, v in counts.items():
    harry_freq[k] = v
```

```python
# сортируем по частоте
harry_sorted = sorted(harry_freq.items(), key=lambda x: x[1], reverse=True)
harry_sorted[:10]
```

```python
# проверяем закон Ципфа
X = range(50)
Y = []

# достаём абсолютную частоту из словаря
             
for word in harry_sorted:
    harry_abs_freq = word[1] 
    Y.append(harry_abs_freq)
             
plt.plot(X, Y[:50], c='xkcd:blue', linestyle='--', linewidth=5)
plt.title('Закон Ципфа')
plt.xlabel('Номер слова')
plt.ylabel('Абсолютная частота')
plt.show()
```

3.	Очистите текст от стоп-слов. Постройте barplot для 30 самых частотных слов. Эти слова должны быть по оси Х в виде подписей к делениям. Значениями по этой оси будет ряд чисел от 0 до 30.

```python
# очищаем от стоп-слов
sw = stopwords.words('english')

harry_filtered = [w for w in word_tokenize(harry_lemmatized) if w not in sw]
print(harry_filtered[:20])
```

```python
# сортируем текст без стоп-слов
harry_nosw_freq = {}
counts = Counter(harry_filtered)

for k, v in counts.items():
    harry_nosw_freq[k] = v
    

harry_nosw_sorted = sorted(harry_nosw_freq.items(), key=lambda x: x[1], reverse=True)
print(harry_nosw_sorted[:30])
```

```python
# строим barplot для 30 самых частотных слов


X2 = range(30)
Y2 = [word[1] for word in harry_nosw_sorted] # достаем из словаря число
Z2 = [word[0] for word in harry_nosw_sorted] # достаем слово для подписи делений


plt.bar(X2, Y2[:30])
plt.title('30 самых частотных слов')
plt.ylabel('Абсолютная частота слова')
plt.xticks(X2, Z2[:30], rotation=90)
plt.show()
```
